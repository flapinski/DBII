```python
import pandas as pd
import numpy as np
```

```python
data = np.array([2369, 3760, 4692, 4871, 5659, 1821, 1074, 7115, 1620, 2428, 3943, 4750, 6975, 4981, 9208])
df = pd.DataFrame({'key': data})
df['hkey'] = np.mod(data, 32)

bit = []
for i in range(len(df['hkey'])):
  bit.append(bin(df['hkey'][i]))

df['bit'] = bit

df
```
